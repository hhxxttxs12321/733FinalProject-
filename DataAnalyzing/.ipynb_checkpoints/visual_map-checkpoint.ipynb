```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib
from shapely.geometry import Point, Polygon
```

```python
flights_df = pd.read_csv('winter_2021_clean.csv')
flights_df = flights_df[['flight_date','departure_icao','arrival_icao']]
flights_df.dropna()
```

```python
airport_df = pd.read_csv('airport.csv')
airport_df = airport_df[['icao_code','country_iso2','latitude','longitude','country_name']]
airport_df.dropna()
```

```python
df_dep = flights_df.merge(airport_df,left_on='departure_icao', right_on='icao_code')
df_dep = df_dep.groupby(['latitude','longitude']).agg(num_flights=('icao_code','size')).reset_index()
df_dep.sort_values(by='num_flights',ascending=False)
df_dep['log_scale_flights']=np.log2(1+df_dep['num_flights'])
```

# Air traffic Density Analysis

```python
df_worldmap = gpd.read_file('./world_map/ne_110m_admin_0_countries.shp')
df_worldmap['coords'] = df_worldmap['geometry'].apply(lambda x: x.representative_point().coords[:])
df_worldmap['coords'] = [coords[0] for coords in df_worldmap['coords']]

crs={'init':'epsg:4326'}
fig, ax = plt.subplots(1, 1,figsize=(40,20))
ax.set_title('Airport Traffic Density Map (Departure)',fontdict={'fontsize':25, 'fontweight': 'medium'})
df_worldmap.plot( ax=ax, legend=True,color = 'lightgrey',edgecolor='grey')
for idx, row in df_worldmap.iterrows():
    if row['LABELRANK'] <= 3:
        plt.annotate(s=row['ADMIN'], xy=row['coords'], horizontalalignment='center',fontsize=10)
geometry=[Point(xy) for xy in zip(df_dep["longitude"], df_dep["latitude"])]
geodata=gpd.GeoDataFrame(df_dep,crs=crs, geometry=geometry)
geodata.plot(column='num_flights',ax=ax,markersize=df_dep['log_scale_flights']*5 ,norm=matplotlib.colors.LogNorm(vmin=df_dep['num_flights'].min(), vmax=df_dep['num_flights'].max())
             ,cmap='autumn_r',legend=True, alpha=.7, legend_kwds={'label': "num_flights", 'shrink': 0.8})
plt.axis('off')
plt.show()
```

```python
df_arr = flights_df.merge(airport_df,left_on='arrival_icao', right_on='icao_code')
df_arr = df_arr.groupby(['latitude','longitude']).agg(num_flights=('icao_code','size')).reset_index()
df_arr.sort_values(by='num_flights',ascending=False)
df_arr['log_scale_flights']=np.log2(1+df_arr['num_flights'])
```

```python
crs={'init':'epsg:4326'}
fig, ax = plt.subplots(1, 1,figsize=(40,20))
ax.set_title('Airport Traffic Density Map (Arrival)',fontdict={'fontsize':25, 'fontweight': 'medium'})
df_worldmap.plot( ax=ax, legend=True,color = 'lightgrey',edgecolor='grey')
for idx, row in df_worldmap.iterrows():
    if row['LABELRANK'] <= 3:
        plt.annotate(s=row['ADMIN'], xy=row['coords'], horizontalalignment='center',fontsize=10)
geometry=[Point(xy) for xy in zip(df_arr["longitude"], df_arr["latitude"])]
geodata=gpd.GeoDataFrame(df_arr,crs=crs, geometry=geometry)
geodata.plot(column='num_flights',ax=ax,markersize=df_arr['log_scale_flights']*5 ,norm=matplotlib.colors.LogNorm(vmin=df_arr['num_flights'].min(), vmax=df_arr['num_flights'].max())
             ,cmap='autumn_r',legend=True, alpha=.7, legend_kwds={'label': "num_flights",'shrink': 0.8})
plt.axis('off')
plt.show()
```

```python
df_dep = flights_df.merge(airport_df,left_on='departure_icao', right_on='icao_code')
df_dep = df_dep.groupby(['country_iso2']).agg(num_flights=('icao_code','size')).reset_index()
df_dep=df_dep.sort_values(by='num_flights',ascending=False).head(10)
fig, ax = plt.subplots(1, 1)
ax.set_title('The top 10 countries that have most flights departure',fontsize=15)
ax.set_xlabel('Country Name')
ax.set_ylabel('Flights Number')
plt.bar(df_dep['country_iso2'],df_dep['num_flights'])
#plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')
plt.show()
```

```python
df_arr = flights_df.merge(airport_df,left_on='arrival_icao', right_on='icao_code')
df_arr = df_arr.groupby(['country_iso2']).agg(num_flights=('icao_code','size')).reset_index()
df_arr = df_arr.sort_values(by='num_flights',ascending=False).head(10)
fig, ax = plt.subplots(1, 1)
ax.set_title('The top 10 countries that have most flights arrival',fontsize=15)
ax.set_xlabel('Country Name')
ax.set_ylabel('Flights Number')
plt.bar(df_arr['country_iso2'],df_arr['num_flights'])
#plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')
plt.show()
```

```python
flights_df_country = flights_df.merge(airport_df,left_on='arrival_icao', right_on='icao_code')
flights_df_country = flights_df_country.merge(airport_df,left_on='departure_icao', right_on='icao_code',suffixes=('_arrival', '_departure'))
flights_df_country = flights_df_country[['flight_date', 'arrival_icao','country_iso2_arrival','country_name_arrival','departure_icao','country_iso2_departure','country_name_departure']]
flights_df_country = flights_df_country[flights_df_country['country_iso2_arrival'] !=flights_df_country['country_iso2_departure']]
flights_df_country
```

```python
df_between_country= flights_df_country.groupby(['country_iso2_arrival', 'country_iso2_departure']).agg(num_flights=('arrival_icao','size')).reset_index()
df_between_country = df_between_country.sort_values(by='num_flights',ascending=False)

df_between_country.to_csv('flights_between_countries.csv')
```

```python

import graph_tool.all as gt
import pandas as pd
import matplotlib.pyplot as plt
countries_df = pd.read_csv('./flights_between_countries.csv')
countries_df['arr_dep']=countries_df['country_iso2_arrival']+countries_df['country_iso2_departure']
countries_df['dep_arr']=countries_df['country_iso2_departure']+countries_df['country_iso2_arrival']
df_between = countries_df.merge(countries_df, left_on='arr_dep',right_on='dep_arr',suffixes=('', '_x'))
df_between['total_flights'] = df_between['num_flights']+df_between['num_flights_x']
df_between = df_between[df_between['total_flights']>100]
df_between_list = df_between[['country_iso2_arrival', 'country_iso2_departure']].values.tolist()

#df_between.sort_values(by='total_flights',ascending=False)
for i in range(len(df_between_list)):
    df_between_list[i].sort()
    df_between_list[i] = tuple(df_between_list[i])
df_between_list = set(df_between_list)
df_between_list
```

```python
g = gt.Graph(directed=False)
name = g.add_edge_list(((n,k) for n,k in df_between_list),hashed=True)
gt.remove_parallel_edges(g)

pos = gt.sfdp_layout(g)
#vprops = dict(font_size=5, fill_color='blue', pen_width=2)
vprops = dict(
    # TODO find suitable vertex properties for graph_draw (see gt reference docs)
    color='dimgray',
    fill_color='#f0f0e0',
    text=name,
    font_size=10,
    pen_width=2
    
)
print('Vountries with more than 100 flights with each other')
gt.graph_draw(    g,
    pos=pos,
    vprops=vprops,
    output_size=(500,500), # adjust to fit screen
    fit_view=.9,
    fmt='svg')
plt.show()
```
